```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.metrics import confusion_matrix
data = pd.read_excel("/content/clustering data.xlsx")
#dataset= data[["Frequency", "Money_spent","Time_spent"]]
linkage_dataset1 = linkage(data, method='average', metric='euclidean')
dendrogram(linkage_dataset1)
```

```python
linkage_dataset2 = linkage(data, method='complete', metric='euclidean')
dendrogram(linkage_dataset2)

from sklearn.cluster import AgglomerativeClustering
hierarchical_cluster = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='complete')
predictlabels2 = hierarchical_cluster.fit_predict(data)
print(predictlabels2)
np.mean(data[predictlabels2 == 1])
```
